```python
import pandas as a
#variabel
```

```python
d = {'pemain' : ['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'joni'],
     'goal' : [2, 1, 3, 1, 1, 2, 2, 3]}
df = a.DataFrame(d)
df
#buat tabel
```

```python
df['goal'].cumsum().to_frame()
#menambahkan angka dengan angka setelahnya/ kumulatif
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
#menambahkan tabel dengan isi angka yang ditambahkan dengan angka setelahnya/ kumulatif
```

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
#menambahkan tabel lagi dengan isi angka yang ditambahkan dengan angka setelahnya tetapi untuk setuiap pemain saja
```

```python
df['cummax'] = df['goal'].cummax()
df
#membuat tabel baru dengan isi angka hasil kumulatif cummax
```

```python
df['cummax'] = df['goal'].cummin()
df
#membuat tabel baru dengan isi angka hasil kumulatif cummin
```

```python
df['cummax'] = df['goal'].cumprod()
df
#membuat tabel baru dengan isi angka hasil kumulatif cumprod
```

```python
df = a.DataFrame({'jenis_kelamin' : ['Pria', 'Wanita', 'lelaki', 'lelaki', 'perempuan'],
                  'usia' : [23, 21, 24, 22, 21],
                  'shift' : ['pagi', 'siang', 'malam', 'siang', 'pagi']})
df
#membuat tabel
```

```python
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
#mapping menggunakan applymap
```

```python
mapping = {'pria' : 'L',
           'lelaki' : 'L',
           'wanita' : 'P',
           'perempuan' : 'P',
           'pagi' : 1,
           'siang' : 2,
           'malam' : 3}
df.applymap(mapping.get)
#membuat dictionary dari applymap
```

```python
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)
df
```

```python
d = {'no_nota' : [1, 1, 1, 2, 2, 3, 4, 5],
     'kopi' : ['latte', 'capuccino', 'espresso', 'latte', 'espresso', 'capuccino', 'latte', 'espresso'],
     'harga' : [50, 60, 80, 150, 120, 60, 100, 40]}
df = a.DataFrame(d)
df
#buat tabel
```

```python
df.groupby('no_nota')['harga'].sum().to_frame()
#menghitung total angka di setiap angka yg ditentukan di kolom 'no_nota'
```

```python
df['total_harga'] = df.groupby('no_nota')['harga'].transform(sum)
df
#menambahkan tabel berisi angka total di setiap angka di kolom 'no_nota'
```

```python
df.groupby('kopi')['harga'].sum().to_frame
#menghitung jumlah angka dis etiap variabel di kolom kopi
```

```python
df['total_omset'] = df.groupby('kopi')['harga'].transform(sum)
df
#menambahkan tabel baru berisi angka total di setiap variabel di kopi
```

```python
data ={'nama' : ['bayu', 'indra', 'devi', 'agni'],
       'jenis_kelamin' : ['L', 'L', 'P', 'L'],
       'usia' : [23, 21, 22, 25]}
df = a.DataFrame(data)
df
#buat tabel
```

```python
df['nama'].str.cat(df['jenis_kelamin'], sep = ', ').to_frame()
#menyatukan 2 kolom menjadi 1
```

```python
df['nama'].str.cat(df['usia'].astype('str'), sep = ' - ').to_frame()
#menggabung 2 tabel menjadi 1
```

```python
df['nama_usia'] = df['nama'].str.cat(df['usia'].astype('str'), sep = ' - ').to_frame()
df
#membuat kolom baru berisi variabel gabungan dari kolom nama dan usia
```
