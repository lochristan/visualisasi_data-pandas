```python
import pandas as a #variabel
import numpy as b
```

```python
n_row = 10
n_col = 1
col = ('usia',)
df = a.DataFrame(b.random.randint(1,99,size=(n_row, n_col)),
                columns=col)
df
#buat tabel 10 baris 1 kolom namanya usia isinya angka acak 1 - 99
```

```python
df['kelompok_usia'] = a.cut(df['usia'],
                            bins = [0,18,65,99],
                            labels = ['anak', 'dewasa', 'manula'])
df
#membuat kategori umur anak 0 - 18, dewasa 19 - 65, manula 66 - 99 berdasarkan tabel yang sudah dibuat diatas
```

```python
n_row = 5
n_col = 5
col = tuple('ABCDE')
df = a.DataFrame(b.random.randint(1,20,size=(n_row,n_col)),
                 columns=col)
df.head()
#buat tabel 5 kolom namanya abcde dan 5 baris. isinya angka random 1 - 20
```

```python
df1 = df.copy(deep = True)
df1 = df1.drop([1, 4])
df1
#perintah untuk menghilangkan tabel yang mau kita hilangkan
```

```python
df2 = df.copy(deep = True)
df2 = df2.drop([0, 3])
df2
#perintah untuk menghilangkan tabel yang mau kita hilangkan
```

```python
df_inner = a.merge(df1, df2, how = 'inner')
df_inner
#menggabungkan 2 buah tabel denggan metode inner. jadi mengambil baris dengan angka yang sama
```

```python
df_outer = a.merge(df1, df2, how = 'outer')
df_outer
#menggabungkan 2 buah tabel denggan metode outer. jadi ambil semua baris dari df1
#tapi kalo di df2 ada baris yang isinya sama persis bakal diilangkan
```

```python
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'],
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = a.DataFrame(data)
df
#buat tabel, 2 kolom yang satu namanya "nama" dan satu lagi namanya "tempat_kelahiran" dengan isi yang sudah dicantum
```

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
#membuat 2 kolom baru dengan nama 'nama_depan' dan 'nama_belakang'. isinya diambil dari data 'nama' dengan misahin berdasarkan spasi
```

```python
df[['kota', 'provinsi']] = df['tempat_kelahiran'].str.split(', ', expand=True)
df
#sama kaya diatas cuman yang ngebedain yang jadi pemisahnya berdasarkan koma
```
