```python
n_row = 5 #membuat tabel dengan pembuatan tanggal 
n_col = 2
col = ['bilangan pecahan', 'bilangan bulat'] #membuat column bilangan pecahan dan bulat

df = lo.DataFrame(uis.random.randint(1, 20, size=(n_row, n_col)),# membuat angka di column yang isinya angka random 1 - 20
                 columns=col)
df['bilangan pecahan'] = df['bilangan pecahan'].astype('float') #mengubah angka di kolumn bilangan pecahan jadi float

df.index = lo.util.testing.makeDateIndex(n_row, freq='l')
df = df.reset_index()

df['text'] = list('VWXYZ') #membuat tabel baru yang namanya text isinya vwxyz

df
```
