```python
n_row = 5
n_col = 2
col = ['omset', 'operasional']
df = a.DataFrame(b.random.randint(1, 20, size = (n_row, n_col)),
                 columns = col)
df
#buat tabel dengan nama kolom yang sudah dicantum dengan isi angka random 1 - 20
```

```python
df['omset'] = df['omset'] * 100_000
df['operasional'] = df['operasional'] * 10_000
df
#mengkali isi data dengan angka yang kita minta
```

```python
df.index = a.util.testing.makeDateIndex(n_row, freq = 'D')
df = df.reset_index()
df = df.rename(columns = {'index' : 'tanggal'})
df
#nambah kolom baru dengan tipe data tanggal 
```

```python
format = {'tanggal' : '{:%d/%m/%y}',
          'operasional' : 'Rp {:.2f}',
          'omset' : 'Rp {:.2f}'}
laporan = df.style.format(format)
laporan
#memformat ulang semua data yang ada
```

```python
type(laporan)
#mengecek tipe data laporan
```

```python
laporan.hide_index()
#menghilangkan index yang angka kaya 01234 diatas
```

```python
laporan.highlight_min('omset', color='pink')
laporan.highlight_max('omset', color='lightgreen')
laporan.highlight_min('operasional', color='lightblue')
laporan.highlight_max('operasional', color='grey')
#mewarnai angka yang kita minta dengan warna yang kita minta
```
